### Notes:

- The bot has been redesigned so that the trunk of the robot serves as the base.
- Each link and joint has specific identifiers, crucial for understanding the bot's structure and movement.
- PyBullet lacks direct link information retrieval, making joint-link mapping challenging.
- The following is the joint-link mapping obtained using `p.getJointInfo`:

#### Joint 0:
- Type: Revolute joint
- Link: 'stand'

#### Joint 1:
- Type: Slider joint
- Links: 'stand' to 'clipper_comp'

#### Joint 2:
- Type: Revolute joint
- Link: 'gripper_front'

#### Joint 3:
- Type: Revolute joint
- Link: 'gripper_2'

#### Additional Details:
- 'clipper_comp' is a composite part allowing hand movement along the stand.
- Front and back grippers are linked to 'clipper_comp'.
